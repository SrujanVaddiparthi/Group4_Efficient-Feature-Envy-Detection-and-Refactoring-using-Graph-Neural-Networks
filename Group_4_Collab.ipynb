```python
for project in df['project'].unique():
    pivot = df[df['project'] == project].pivot(index="epochs", columns="vectorsize", values="f1score")
    plt.figure(figsize=(6, 4))
    sns.heatmap(pivot, annot=True, fmt=".2f", cmap="YlGnBu")
    plt.title(f"{project.upper()} - F1 Score Heatmap")
    plt.xlabel("Vector Size")
    plt.ylabel("Epochs")
    plt.tight_layout()
    plt.show()
```

```python
best_f1_per_project = df.sort_values(by='f1score', ascending=False).groupby('project').first().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(data=best_f1_per_project, x='project', y='f1score', palette='viridis')
plt.title("Best F1 Score per Project")
plt.ylabel("F1 Score")
plt.xlabel("Project")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='accuracy', y='f1score', hue='project', style='vectorsize', size='epochs', palette='deep')
plt.title("Accuracy vs F1 Score across Configurations")
plt.xlabel("Accuracy")
plt.ylabel("F1 Score")
plt.grid(True)
plt.tight_layout()
plt.show()
```
